```python
# Define the curve
y_curve = 0.4*(x-2)*(x-3)*(x-5)*(x-1)

# Plot the curve
fig, ax = plt.subplots()

# Set the background color to white
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

# Remove the axis numbers
ax.set_xticks([])
ax.set_yticks([])

# Set the labels
ax.set_xlabel('x')
ax.set_ylabel(r'$\phi(x)$')


# Remove the borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Plot the curve
ax.plot(x, y_curve, label='Custom Curve')


# Set x-axis limits
ax.set_xlim(0, 6)
ax.set_ylim(-5, 10)
# Add a legend


ax.plot(1.39275, -0.5531, 'ro')  # Local minimum
ax.plot(4.32635, -2.76564, 'bo')  # Global minimum

# Add labels
ax.annotate('Local Minimum', xy=(1.39275, -0.5531), xytext=(2, 2),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('Global Minimum', xy=(4.32635, -2.76564), xytext=(2, -4),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```
