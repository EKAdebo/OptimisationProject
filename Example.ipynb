### Example 1

MINIMISE 
$\\f_1(x) = 10x_1 + 2x_2\\$
$f_2(x) = 3x_1 ^ 2 + 2x_2 ^ 2 + x_1$ 

SUBJECT TO
$\\ x_1 \geq 2\\$
$ x_2 \geq 3 \\$
$ x_1 + x_2 \leq 12$

```python
#Define the objective functions
m = ["f1","f2"]

#f1
def f1(x):
    return 10*x[0] + 2*x[1]

#Partial derivatives of f1
def f1_d(x):
    return [10,2]

#Cretae object of class Function
func_f1 = Function(f1,f1_d,"f1")

def f2(x):
    return 3* x[0]**2 + 2*x[1]**2 + x[0]

def f2_d(x):
    return [6*x[0] + 1 ,4*x[1]]

func_f2 = Function(f2,f2_d,"f2")

functions = [func_f1,func_f2]

#Define the inequality constraints
def g1(x):
    return 2 - x[0]

def g1_d(x):
    return [-1,0]

constraint_g1 = Constraint(g1,g1_d,"g1")

def g2(x):
    return 3 - x[1]

def g2_d(x):
    return [0,-1]

constraint_g2 = Constraint(g2,g2_d,"g2")

def g3(x):
    return x[1] + x[0] - 12

def g3_d(x):
    return [1,1]

constraint_g3 = Constraint(g3,g3_d,"g3")

#Define a list of the functions
functions = [func_f1,func_f2]
g = [constraint_g1,constraint_g2,constraint_g3]
h = []
```

```python

np.random.seed(41)

n = 100
iterations = 50
originalPoints, nonDominated = SQPMethod(functions, g, h, 2, n, iterations, mu, beta)
```

### Example Problem 3

MINIMISE 
$\\f_1(x) = (x_1 - 2)^2 + (x_2 -3)^2\\$
$f_2(x) = (\frac{x_1}{2})^2 +(\frac{x_2}{5})^2 + 10 sin(x_1)cos(x_2) $

SUBJECT TO
$\\1 \le x_1 \le 5\\$
$0 \le x_" \le 6\\$


```python
#STEP 2
# Create 3 objects of class Point with random vectors of dimension 3


from math import cos, sin


m = ["f1","f2"]

#f1
def f1(x):
    return (x[0] - 2)**2 + (x[1] - 3)**2

#Partial derivatives of f1
def f1_d(x):
    return [2*(x[0] - 2),2*(x[1] - 3)]

#Cretae object of class Function
func_f1 = Function(f1,f1_d,"f1")

def f2(x):
    return (x[0]/2)**2 + (x[1]/5)**2 + 10 * sin(x[0])*cos(x[1])

def f2_d(x):
    t1 = x[0]/2 + 10*cos(x[0])*cos(x[1])
    t2 = (2 *x[1])/25 - 10*sin(x[0])*sin(x[1])
    return [t1,t2]

func_f2 = Function(f2,f2_d,"f2")



def g1(x):
    return 1 - x[0]

def g1_d(x):
    return [-1,0]

constraint_g1 = Constraint(g1,g1_d,"g1")

def g2(x):
    return x[0] - 5

def g2_d(x):
    return [1,0]

constraint_g2 = Constraint(g2,g2_d,"g2")

def g3(x):
    return -x[1]

def g3_d(x):
    return [0,-1]

constraint_g3 = Constraint(g3,g3_d,"g3")

def g4(x):
    return x[1] - 6

def g4_d(x):
    return [0,1]

constraint_g4 = Constraint(g4,g4_d,"g4")


functions = [func_f1,func_f2]
g = [constraint_g1,constraint_g2,constraint_g3,constraint_g4]
h = []
```

```python

np.random.seed(41)

n = 500
iterations = 40
originalPoints, nonDominated = SQPMethod(functions, g, h, 2, n, iterations, mu, beta)
```
